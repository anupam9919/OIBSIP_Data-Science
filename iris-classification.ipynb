```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
coloumns=['SepalLengthCm','SepalWidCm','PetalLengthCm','PetalWidthCm','Species']

d_df=pd.read_csv('/kaggle/input/iriscsv/Iris.csv')
df = d_df[d_df.columns.difference(['Id'])]
df.head()
```

```python
#using seaborn library
sns.pairplot(df,hue='Species')
```

```python
# Separate features and target
data= df.values
X=data[:,0:4]
Y=data[:,4]
print(X)
```

```python
 # Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
print ()
```

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression(solver='lbfgs', max_iter=1000)
model_lr.fit(X_train,Y_train)
```

```python
prediction1= model_lr.predict(X_test)
#Calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1)*100)
for i in range(len(prediction1)):
    print(Y_test[i],prediction1[i])
```

```python
# detailed classification report

from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction1))
```
